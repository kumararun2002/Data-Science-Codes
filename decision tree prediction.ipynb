# Prediction using Decision Tree

## Input Dataset

```python
import numpy as np
from sklearn import datasets
import requests
#import csv

iris_data=datasets.load_iris()
#print(iris_data.data)

X=iris_data.data
Y=iris_data.target

#print(Y)
```

## Splitting of data into Train and Test Set randomly

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,random_state=0)

print(X_test)
print(Y_test)
```

## Applying Decision Tree Classifier on Train Set 

```python
from sklearn import tree

tree_classfier=tree.DecisionTreeClassifier()
tree_classfier.fit(X_train,Y_train)

print("Decision Tree Classifier created successfully!!!")
```

## Visualising the created Tree

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (15,12)

tree.plot_tree(tree_classfier,filled=True) 
plt.show()
```

## Prediction on random value and Test Set

```python
# Let's take a value near to X[0] so prediction should be 0.
test=np.array([5.0,3.3,1.4,0.2])
print(tree_classfier.predict([test]))
```

#### It predicted 0. So my code is predicting correct.

```python
# Let's predict on test set
Pred_Y=tree_classfier.predict(X_test)
print(Pred_Y)
```

## Evaluation

```python
# Let's check the accuracy on test set prediction.
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, Pred_Y))
```

#### Accuracy of 97.78% on test set.

## Getting the confusion matrix

```python
# Let's print the confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test, Pred_Y))
```

## Calculating Precision, Recall and F1-Score

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

print(precision_score(Y_test, Pred_Y, average='weighted'))
print(recall_score(Y_test, Pred_Y, average='weighted'))
print(f1_score(Y_test, Pred_Y, average='weighted'))
```
